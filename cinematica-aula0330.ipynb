# Fórmula da velocidade
fonte: [https://www.todamateria.com.br/formulas-de-cinematica/]

```
def calculaVelMedia(deltaS, deltaT):
  velMedia = (deltaS/deltaT)
  return velMedia
```

## Teste Serviço Nuvem: Velocidade Média

```
print("Vm=", calculaVelMedia(30, 15))
```

# Fórmula da aceleração

```
def calculaAceMedia(veloF, veloI, temF, temI ):
  aMedia = (veloF - veloI/temF - temI)
  return aMedia
```

## Teste Serviço Nuvem: Aceleração Média

```
print("Am=", calculaAceMedia(4, 2, 8, 12))
```

# Fórmulas do Movimento Uniforme (MU)

```
def calculaMU(S, So, v, t):
  S = (So + v*t)
  return S
```

## Teste Serviço Nuvem: Movimento Uniforme

```
print("S=", calculaMU(4, 2, 8, 18))
```

# Função horária do deslocamento

```
def calculaDES(S, So, v, deltat):
  S = (So + v*deltat)
  return S
```

```
print("Deslocamento=", calculaDES(3, 5, 9, 19))
```

# Fórmulas do Movimento Uniformemente Variado (MUV)

```
def calculaMUV(V, Vo, a, t):
  V = (Vo + a*t)
  return V
```

## Teste Serviço Nuvem: Movimento Uniforme Variado

```
print("V=", calculaMUV(4, 2, 8, 18))
```

## Função horária da velocidade

```
def calculaFVEl(V, Vo, a, deltat):
  V = (Vo + a*deltat)
  return V
```

## Teste Serviço Nuvem: Função Horária da Velocidade

```
print("FunçãoVelocidade=", calculaFVEl(4, 2, 8, 18))
```

## Função horária da posição

```
def calculaFPoS(s, so, Vo, a, t):
  s = (so + Vo*t + 1/2 a*t **2)
  return s
```

```
print("FunçãoPosição=", calculaFPoS(4, 2, 8, 18))
```
